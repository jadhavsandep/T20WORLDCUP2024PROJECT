```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("ICC Mens T20 Worldcup.csv")  # Replace 'your_file.csv' with your actual file name
data.head()
```

```python
# Overview of the dataset
data.info()
```

```python
# Check for missing values in each column
data.isnull().sum()
```

```python
# Summary statistics for numerical columns
data.describe()
```

```python
# Display column names
data.columns
```

```python
# Group by players and calculate total runs
top_players = data.groupby('Player')['Runs'].sum().nlargest(10)

# Plot top 10 players by runs
top_players.plot(kind='bar', figsize=(10, 6), title='Top 10 Players by Total Runs', color='skyblue')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.show()
```
