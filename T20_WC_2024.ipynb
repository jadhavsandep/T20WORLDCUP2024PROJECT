```python
data['Winners'].value_counts().plot(kind='bar', title='Wins by Teams', figsize=(10, 5))
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.show()
```

```python
data['Toss Decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Toss Decisions', figsize=(6, 6))
plt.show()
```

```python
top_scorers = data[['Top Scorer', 'Highest Score']].sort_values(by='Highest Score', ascending=False).head(10)
plt.bar(top_scorers['Top Scorer'], top_scorers['Highest Score'])
plt.title('Top Scorers in the Tournament')
plt.xlabel('Players')
plt.ylabel('Scores')
plt.xticks(rotation=45)
plt.show()
```

```python
# Simple bar chart for winning margins
data['Won by'].value_counts().plot(kind='bar', title='Winning Margins by Match Type')
plt.xlabel('Match Result Type')
plt.ylabel('Number of Matches')
plt.show()
```

```python
# Simple pie chart for toss decisions
data['Toss Decision'].value_counts().plot(kind='pie', autopct='%1.1f%%', title='Toss Decisions')
plt.ylabel('')  # Remove the y-axis label for clarity
plt.show()
```

```python
# Top 10 scorers by highest score
top_scorers = data[['Top Scorer', 'Highest Score']].sort_values(by='Highest Score', ascending=False).head(10)
print(top_scorers)
```

```python
# Top 10 bowlers by wickets taken
top_bowlers = data[['Best Bowler', 'Best Bowler Figure(Wickets Taken)']].sort_values(by='Best Bowler Figure(Wickets Taken)', ascending=False).head(10)
print(top_bowlers)
```

```python
# Top 10 players by Player of the Match awards
top_players = data['Player Of The Match'].value_counts().head(10)
print(top_players)
```

```python
# Bar chart for top scorers
top_scorers = data[['Top Scorer', 'Highest Score']].sort_values(by='Highest Score', ascending=False).head(10)
plt.bar(top_scorers['Top Scorer'], top_scorers['Highest Score'])
plt.title('Top 10 Scorers by Highest Score')
plt.xlabel('Player')
plt.ylabel('Highest Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart for top bowlers
top_bowlers = data[['Best Bowler', 'Best Bowler Figure(Wickets Taken)']].sort_values(by='Best Bowler Figure(Wickets Taken)', ascending=False).head(10)
plt.bar(top_bowlers['Best Bowler'], top_bowlers['Best Bowler Figure(Wickets Taken)'])
plt.title('Top 10 Bowlers by Wickets Taken')
plt.xlabel('Bowler')
plt.ylabel('Wickets Taken')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart for Player of the Match awards
top_players = data['Player Of The Match'].value_counts().head(10)
top_players.plot(kind='bar', title='Top 10 Players by Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.xticks(rotation=45)
plt.show()
```
